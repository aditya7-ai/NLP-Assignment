```python
import nltk
from nltk.corpus import wordnet as wn

# Download WordNet (run once)
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
word = "happy"
synsets = wn.synsets(word)

print("Synsets of the word:", word)
for syn in synsets:
    print(syn.name(), "-", syn.definition())
```

```python
# synonyms
synonyms = set()

for syn in synsets:
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms:")
print(synonyms)
```

```python
# antonyms
antonyms = set()

for syn in synsets:
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms:")
print(antonyms)
```

```python
# hypernyms
word = "dog"
synsets = wn.synsets(word)

hypernyms = set()

for syn in synsets:
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name())

print("Hypernyms:")
print(hypernyms)
```

```python
# hyponyms
word = "car"
synsets = wn.synsets(word)
hyponyms = set()

for syn in synsets:
    for hypo in syn.hyponyms():
        hyponyms.add(hypo.name())

print("Hyponyms:")
print(hyponyms)
```
